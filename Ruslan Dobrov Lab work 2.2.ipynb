### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random

#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(6,100) for i in range(1000)]


#рохраховує мінімальне та максимальне значення
a = min(list_with_random_numbers)
b = max(list_with_random_numbers)
print('Мінімальне та максимальне значення -', a,'та',b)


# середнє значення послідовності
average = sum(list_with_random_numbers) / 1000
print('Середнє значення -', average)


# медіану послідовності
list_with_random_numbers.sort()
if len(list_with_random_numbers) % 2 == 0:
    i1 = len(list_with_random_numbers) // 2
    i2 = len(list_with_random_numbers) // 2 - 1
    print('Медіана послідовності -',(list_with_random_numbers[i1]+ list_with_random_numbers[i2])/2)
else:
    i = len(list_with_random_numbers) // 2
    print('Медіана послідовності -', list_with_random_numbers[i])
    
    
# частоту кожної цифри в послідовності
one_time_numbers = []
counter_list = []
a = 0
for i in list_with_random_numbers:
    if i not in one_time_numbers:
        one_time_numbers.append(i)
        counter_list.append(list_with_random_numbers.count(i))
final_list = list(zip(one_time_numbers, counter_list))
# здесь по парам разбиты (число из начального списка, кол-во раз встречается в списке)
print(final_list)
```

```python
# вот второй вариант разбивания на пары.
one_time_numbers = []
counter_list = []
a = 0
for i in list_with_random_numbers:
    if i not in one_time_numbers:
        one_time_numbers.append(i)
        counter_list.append(list_with_random_numbers.count(i))
final_list = list(zip(one_time_numbers, counter_list))

for i in final_list:
    print('Число', i[0], 'встречается', i[1], 'раз')
```

```python
# Середнє квадратичне відхилення
# Вычисляем среднее арифметическое. Нам пригодится в дальнейших расчетах.
summ = 0
for i in list_with_random_numbers:
    summ += i
average = summ / 1000
#print('Среднее арифметическое =', average)

# Определяем отклонение каждого элемента относительно среднего арифметического
standard_deviation_for_i = []
for i in list_with_random_numbers:
    x = i - average
    standard_deviation_for_i.append(x)
    
# Определим квадрат разницы отклонения значений ряда относительно среднего
squared_stand_dev = []
for i in standard_deviation_for_i:
    squared = i ** 2
    squared_stand_dev.append(squared)
    
# Рассчитаем сумму квадратов отклонений значений относительно среднего
summary_squares = sum(squared_stand_dev)
#print('Сумма квадратов отклонений =',summary_squares)

# Делим сумму квадратов отклонений на кол-во элементов минус 1
dev_for_i = summary_squares / (len(list_with_random_numbers) - 1)

# Среднеквадратическое отклонение собственно
import math
standart_deviation = math.sqrt(dev_for_i)
print('Среднеквадратическое отклонение =', standart_deviation)
```
